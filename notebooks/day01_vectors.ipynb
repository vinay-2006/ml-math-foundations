## Why Vectors?

Traditional data analysis treats data as **rows in a table**.
Machine Learning treats data as **points in space**.

| Perspective | Interpretation |
|------------|----------------|
| Spreadsheet | Rows & columns |
| ML Model | Geometry in $n$-dimensional space |

From this point forward:

> **One data sample = one vector**


```python
import math
import matplotlib.pyplot as plt
```

## Scalars vs Vectors

### Scalar
A scalar is a **single number**.

Examples:
- Temperature
- Learning rate
- Loss value

### Vector
A vector is an **ordered collection of numbers**.

Each element corresponds to a **feature**.

Example:

$$
x = [\text{hours studied},\ \text{sleep hours},\ \text{previous score}]
$$

Machine Learning models operate on **vectors**, not scalars.


```python
student_1 = [6, 7, 78]
student_2 = [4, 6, 65]

student_1, student_2
```

## Vector as a Data Point

A vector is not just a list — it is a **point in space**.

- 2 features → 2D space  
- 3 features → 3D space  
- $n$ features → $n$-dimensional space  

ML models reason geometrically in $\mathbb{R}^n$.


## Vector Magnitude (Norm)

Magnitude answers the question:

> **How far is this point from the origin?**

For a vector:

$$
x = [x_1, x_2, \dots, x_n]
$$

The Euclidean norm is:

$$
\|x\| = \sqrt{x_1^2 + x_2^2 + \dots + x_n^2}
$$

Magnitude reflects the **scale** of a data point.


```python
def magnitude(vector):
    return math.sqrt(sum(v ** 2 for v in vector))


magnitude(student_1), magnitude(student_2)
```

## Distance Between Vectors

Distance measures **similarity** between samples.

- Small distance → similar samples  
- Large distance → dissimilar samples  

The Euclidean distance between $x$ and $y$ is:

$$
d(x, y) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}
$$

Distance is the backbone of similarity-based ML methods.


```python
def euclidean_distance(x, y):
    return math.sqrt(sum((a - b) ** 2 for a, b in zip(x, y)))


euclidean_distance(student_1, student_2)
```

## Vector Arithmetic

Machine Learning does not only *measure* vectors — it **updates and moves them**.

The two operations used throughout ML are:
1. Vector addition / subtraction
2. Scalar multiplication

Without these, Gradient Descent cannot be defined.


### Vector Addition and Subtraction

Given vectors:

$$
x = [x_1, x_2, \dots, x_n], \quad
y = [y_1, y_2, \dots, y_n]
$$

Addition:

$$
x + y = [x_1 + y_1,\ \dots,\ x_n + y_n]
$$

Subtraction:

$$
x - y = [x_1 - y_1,\ \dots,\ x_n - y_n]
$$

**ML Meaning**

- Updating parameters is vector subtraction
- Learning is movement in parameter space


```python
def vector_add(x, y):
    return [a + b for a, b in zip(x, y)]

def vector_subtract(x, y):
    return [a - b for a, b in zip(x, y)]


v1 = [2, 3]
v2 = [1, 4]

vector_add(v1, v2), vector_subtract(v1, v2)
```

### Scalar Multiplication

A scalar changes a vector’s **magnitude**, not its direction.

For scalar $\eta$ and vector $x$:

$$
\eta x = [\eta x_1,\ \eta x_2,\ \dots,\ \eta x_n]
$$

**ML Meaning**

- $\eta$ is the learning rate
- It controls step size during optimization


```python
def scalar_multiply(scalar, vector):
    return [scalar * v for v in vector]


learning_rate = 0.1
gradient = [2.0, -3.0]

scalar_multiply(learning_rate, gradient)
```

## Preview: Why This Matters

Gradient Descent update rule:

$$
w = w - \eta \nabla w
$$

This involves:
- Scalar multiplication → $\eta \nabla w$
- Vector subtraction → $w - (\cdot)$

## Visualizing Vectors in 2D

To strengthen geometric intuition, we visualize vectors as points in 2D space.


```python
points = {
    "A": [2, 3],
    "B": [5, 7],
    "C": [6, 2]
}

xs = [v[0] for v in points.values()]
ys = [v[1] for v in points.values()]

plt.figure(figsize=(6, 6))
plt.scatter(xs, ys)

for label, (x, y) in points.items():
    plt.text(x + 0.1, y + 0.1, label)

plt.axhline(0)
plt.axvline(0)
plt.grid(True)
plt.title("2D Vector Representation of Data Samples")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
